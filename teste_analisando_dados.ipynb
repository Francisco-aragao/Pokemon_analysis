```python
ANO = 2019
RANK = 15
CARD = "Wondrous Labyrinth ♢"
```

```python
import numpy as np
import pandas as pd
import pysubgroup as ps

df_original = pd.read_csv('./Data/tournaments.csv')
df = df_original.copy()

columns_to_remove = [
    'id_card', 'amount_card', 'price_card', 'combo_type_id', 'name_tournament',
    'combo_type_name', 'name_player', 'category_tournament', 'type_card',
    'month_tournament', 'valid_rotation_at_tournament', 
    'rotation_name', 'year_begin', 'all_time_score', 'country_tournament', 
    'country_player', 'day_tournament','month_begin','day_begin'
]
df.drop(columns=columns_to_remove, inplace=True)
df = df[df['region_tournament'].notna()]
```
